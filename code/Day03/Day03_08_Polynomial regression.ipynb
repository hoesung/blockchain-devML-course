# 다항회귀란?
- 독립변수와 종속변수들 간의 관계를 다차 다항식으로 표현하는 회귀 방식
- 예를 들어, 2개의 독립변수 X1, X2와 종속변수 Y의 관계를 아래와 같은 다항식으로 표현
> $ Y = w_0 + w_1X_1 + w_2X_2 + w_3X_1X_2 + w_4X_1^2 + w_5X_2^2 $

## 사이킷런으로 변환 수행연습

```python
import sklearn.preprocessing as pp
```

preprocessing 모듈에 있는 PolynomialFeatures를 이용하여 특성들을 다항식 형태로 변환하는 객체를 생성

````python
import numpy as np
# 원본 특성 데이터 [[0, 1] [2, 3]] 배열
X = np.arange(4).reshape(2,2)
X
````

````python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2) # 매개변수 degree는 변환하려는 다항식의 차수이며, 기본값은 2
X_poly = poly.fit_transform(X)      # 변환 
print(X_poly)
````

````python
# 데이터 설정 
import numpy as np
import matplotlib.pyplot as plt

X = 6 * np.random.rand(100,1) - 3
y = 0.5 * X ** 2 + 2  + np.random.randn(100,1)

plt.scatter(X,y,color='red')
plt.show()
````

````python
X_train = X
y_train = y
X_test =  np.arange(-3,3,0.01)[:,np.newaxis]
# y_test = 
````

````python
# 단순선형회귀 
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression().fit(X_train,y_train)
linear_pred = linear_model.predict(X_test)

plt.plot(X_test,linear_pred,color = 'blue',linestyle='--')
````

````python
# 2차 다항회귀 
from sklearn.preprocessing import PolynomialFeatures
poly = pp.PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X)
poly_model = LinearRegression().fit(X_train_poly,y_train)


X_test_poly = poly.fit_transform(X_test)
poly_pred = poly_model.predict(X_test_poly)

plt.plot(X_test,poly_pred,color='green')
plt.show()
````

````python
# 7 차 다항회귀 
from sklearn.preprocessing import PolynomialFeatures
poly = pp.PolynomialFeatures(degree=7)
X_train_poly_7 = poly.fit_transform(X)
poly_model_7 = LinearRegression().fit(X_train_poly_7,y_train)


X_test_poly_7 = poly.fit_transform(X_test)
poly_pred_7 = poly_model_7.predict(X_test_poly_7)

plt.plot(X_test,poly_pred_7,color='magenta',linestyle=":")
plt.show()
````

````python
plt.scatter(X,y,color='red')
plt.plot(X_test,linear_pred,color = 'blue',linestyle='--')
plt.plot(X_test,poly_pred,color='green')
plt.plot(X_test,poly_pred_7,color='magenta',linestyle=":")
plt.show()
````
