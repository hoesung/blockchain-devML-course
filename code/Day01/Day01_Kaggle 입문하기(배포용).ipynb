````python
def show_continuous_var(dataset,feature,legend_flag=True,font_size = 12):
    plt.figure(figsize=(12,8))
    plt.title('{} 전체 데이터 분포'.format(feature))
    plt.xlabel('{}'.format(feature),fontsize=12)
    plt.ylabel('밀도(Density)',fontsize=12)
    dataset[feature].value_counts().sort_index().plot(kind='kde')
    plt.show()
    if dataset is train:
        plt.figure(figsize=(12,8))
        plt.title('생존 여부별 {} 분포 그래프'.format(feature),fontsize=font_size)
        plt.xlabel('{} 의 수'.format(feature),fontsize=font_size)
        plt.ylabel('밀도(Density)',fontsize=font_size)
        survived = dataset[dataset['Survived']==1][feature].value_counts() 
        dead = dataset[dataset['Survived']==0][feature].value_counts() 
        dataset[dataset['Survived']==1][feature].value_counts().sort_index().plot(kind='kde',label='Survived')
        dataset[dataset['Survived']==0][feature].value_counts().sort_index().plot(kind='kde',label='Dead')
        plt.legend()
        plt.show()
````

### 범주형 변수 살펴보기

````python
def show_category_var(dataset,feature,legend_flag=True,font_size=12):
    plt.figure(figsize=(12,8))
    plt.title('{} 전체 데이터 분포'.format(feature))
    plt.xlabel('{}'.format(feature),fontsize=12)
    plt.ylabel('갯수',fontsize=12)
    dataset[feature].value_counts().sort_index().plot(kind='bar')
    plt.show()
    if dataset is train:
        survived = dataset[dataset['Survived']==1][feature].value_counts() / len(dataset[dataset['Survived']==1][feature])
        dead = dataset[dataset['Survived']==0][feature].value_counts() / len(dataset[dataset['Survived']==0][feature])
        df = pd.DataFrame([survived,dead])
        df.index = ['Survived','Dead']
        df.plot(kind='bar',stacked=True, figsize=(12,8))
        plt.title('생존 여부별 {} 분포 그래프'.format(feature),fontsize=font_size)
        plt.xlabel('{} 의 수'.format(feature),fontsize=font_size)
        plt.ylabel('비율(%)',fontsize=font_size)
        plt.show()
````

## 피처 엔지니어링(Feature Engineering)

피처 엔지니어링(Feature Engineering)은 머신러닝 모델이 학습을 잘할 수 있게 바꿔주는 일련의 과정들이다.  
여기에는 흔히 말하는 `데이터 전처리(Data Preprocessing)`도 여기에 해당하며, `데이터 스케일링(Data Scaling)`이나 `원핫 인코딩(One Hot Encoding)`등 과 같이 흔히 머신러닝을 공부하면서 들어봤던 기술들을 여기에서 사용한다.

> Kaggle Competition에서 피처 엔지니어링은 전체 과정에 70~80% 이상 할애를 해야하는 작업이다.  
⇒ 사실상 이 과정을 잘하는게 대회에서 고득점을 성취할 수 있는 핵심포인트이다. 귀찮지만 매우 중요한 작업이다.
