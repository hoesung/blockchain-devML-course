```python
import pandas as pd
import numpy as np

raw_data = pd.read_csv('../../data/water.csv', 
                       encoding='CP949', # encoding problem utf-8
                       engine='python')
raw_data.head()
```

```python
# 적합한 데이터만 가져온것
tmp = raw_data[raw_data['적합'] == '적합']
tmp.pivot_table(index = '약수터명',
               aggfunc = 'mean',
               values=['일반세균', '질산성질소'])
```

```python
raw_data = raw_data.interpolate(method='linear', axis=0, limit_direction='both')
```

```python
raw_data
```

```python
raw_data['적합'] = raw_data['적합'].map({'부적합':0 ,'적합':1})
raw_data['총대장균군'] = raw_data['총대장균군'].map({'음 성':0 ,'양 성':1})
```

```python
raw_data
```

```python
X = raw_data[['총대장균군','일반세균','질산성질소']]
y = raw_data['적합']
```

```python
# train_test_split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
                                                   random_state=42)
```

```python
#  모델 학습 fitting
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test,y_test)
```
